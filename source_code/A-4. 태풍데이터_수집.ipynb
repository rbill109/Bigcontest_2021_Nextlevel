```python
def plot_nh2(k, i, df):
    df.index = df.base_date ;
    df.month = df.base_date.dt.month
    dat = df[df.month==8]
#    dat = dat.resample("2D").mean()
    plt.figure(figsize=(20, 5))
    ax1, ax2 = plt.gca(), plt.gca().twinx() 

    sns.lineplot(dat.index,dat[i].values,color = 'darkblue',alpha=1, ax = ax1)
    ax1.set_ylabel('y: {}'.format(i) , color='darkblue', fontsize=15)

    sns.lineplot(dat.index,dat[k].values,color = 'skyblue',alpha=1, ax = ax2)
    ax2.set_ylabel('x: {}'.format(k) , color='skyblue', fontsize=15)

    ax1.set_xlabel("{} & {}".format(i, k) , fontsize=15)
    plt.show()
```

```python
for i in df_78.emd_nm.unique():
    print("==============================================2021년 8월 태풍 루핏, 오마이스 {}===============================================".format(i))
    tmp = df_78.loc[(df_78.emd_nm==i)]
    plot_nh2("기온", "강수", tmp)
```
